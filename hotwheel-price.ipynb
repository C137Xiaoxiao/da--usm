```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_amazon_product_details(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    title = soup.find(id='productTitle')
    price = soup.find('span', {'class': 'a-offscreen'})

    if title and price:
        product_name = title.text.strip()
        product_price = price.text.strip()
        return product_name, product_price
    else:
        return "Product details not found"

# Amazon网页URL
url = 'https://www.amazon.com/Hot-Wheels-Mandalorian-Character-5-Pack/dp/B09NB19SZS/'
product_details = get_amazon_product_details(url)

print(f"Product Name: {product_details[0]}, Price: {product_details[1]}")
```
